```python
df = pd.read_csv("showwcase_sessions.csv")
```

```python
df['total_session_time']=df['session_duration']+ df['inactive_duration']
```

Grouping Customer-wise data

```python
a = df.groupby('customer_id', as_index = False)['session_duration'].sum()
b = df.groupby('customer_id',as_index = False)['inactive_duration'].sum()
a = a.merge(b,how="outer", left_on = ['customer_id'], right_on = ['customer_id'] )
a['active_ratio'] = a['session_duration']/(a['session_duration']+a['inactive_duration'])  #active_ratio=ratio of active session work to total time spent
```

```python
df.columns
```

```python
c = df.groupby('customer_id', as_index=False)['session_projects_added'].sum()
d = df.groupby('customer_id', as_index=False)['session_likes_given'].sum()
e = df.groupby('customer_id', as_index=False)['session_comments_given'].sum()
f = df.groupby('customer_id', as_index=False)['bugs_in_session'].sum()
g = df.groupby('customer_id', as_index=False)['login_date'].nunique()
```

```python
g
```

```python
a = a.merge(c,how="outer", left_on = ['customer_id'], right_on = ['customer_id'] )
a = a.merge(d,how="outer", left_on = ['customer_id'], right_on = ['customer_id'] )
a = a.merge(e,how="outer", left_on = ['customer_id'], right_on = ['customer_id'] )
a = a.merge(f,how="outer", left_on = ['customer_id'], right_on = ['customer_id'] )
```

```python
a['login_frequency_days'] = g
```

```python
a = a.merge(dfx,how="outer", left_on = ['customer_id'], right_on = ['customer_id'] )
```

```python
a = a.rename(columns={'day_of_month': 'Most_Recent_Login_Date'})
```

```python
a.columns
```

Converting Seconds to Minutes

```python
a['session_minutes']=a['session_duration']/60
```

```python
a.head()
```

Scaling all the fields for RFM

```python
a["active_ratio_scaled"] = a["active_ratio"]/max(a["active_ratio"])
a["session_projects_added_scaled"] = a["session_projects_added"]/max(a["session_projects_added"])
```

```python
a["session_likes_given_scaled"] = a["session_likes_given"]/max(a["session_likes_given"])
a["session_comments_given_scaled"] = a["session_comments_given"]/max(a["session_comments_given"])
a["bugs_in_session_scaled"] = a["bugs_in_session"]/max(a["bugs_in_session"])
a["login_frequency_days_scaled"] = a["login_frequency_days"]/max(a["login_frequency_days"])
a["session_minutes_scaled"] = a["session_minutes"]/max(a["session_minutes"])
```

Assigning 0.5 when logging in first 10 days else 1 for next 20 days

```python
a["Most_Recent_Login_Date_scaled"] = a['Most_Recent_Login_Date'].apply(lambda x: 0.5 if x <= 10 else 1)
```

Net Value Of Customer (RFM Score) = Projects + Likes + Comments + Login Frequency + Recent Login Date + Session Minutes

```python
a['Net_value_of_customer']= a['session_projects_added_scaled'] +a['session_likes_given_scaled']+ a['session_comments_given_scaled']+ a['login_frequency_days_scaled'] + a['Most_Recent_Login_Date_scaled']+ a['session_minutes_scaled']
```
